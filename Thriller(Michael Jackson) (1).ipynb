```python
unique_words =set(split_lyrics)
print(unique_words)
```

```python
len(unique_words)
```

```python
len(split_lyrics)
```

```python
# we need to know how many times each word appears so we use the for loop
word_dict = {}
for word in unique_words:
    word_dict[word] = 0
    print(word_dict)
    
```

```python
# To count
for word in split_lyrics:
    word_dict[word] = word_dict[word] + 1
    print(word_dict)
```

```python
# we then sort
sort_dict = sorted(word_dict.items(),key=lambda t:t[1],reverse=True)
print(sort_dict)
```

```python
# slice to top ten list
top_list = sort_dict[0:10]
top_list
```

```python
top10_dict =dict(top_list)
top10_dict
```

```python
uniqueword10=list(top10_dict.keys())
uniqueword10
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

so_axis = np.arange(len(uniqueword10))
plt.title("Thriller",fontsize=20)
plt.xticks(so_axis,uniqueword10)
plt.bar(so_axis,top10_dict.values())
plt.show()
```
