```python
# storing player info in a dictionary and perform some python work then we visualize using matplotlib
# The aim here is to visualize the trend of goals bn scored in the past 5 years in the EPL
players = {
    'player_2015':{
        'first':'aguero',
        'last':'sergio',
        'number_goals':26,
        'country':'argentina',
        'club':'manchester city',
        'year':2015,
        },
    
    'player_2016':{
        'first':'kane',
        'last':'harry',
        'number_goals':25,
        'country':'england',
        'club':'tottenham hotspur',
        'year':2016,
        },
    
    'player_2017':{
        'first':'kane',
        'last':'harry',
        'number_goals':29,
        'country':'england',
        'club':'tottenham hotspur',
        'year':2017,
        },
    
    'player_2018':{
        'first':'salah',
        'last':'mohamed',
        'number_goals':32,
        'country':'egypt',
        'club':'liverpool',
        'year':2018,
        },
    
    'player_2019':{
        'first':'salah',
        'last':'mohamed',
        'number_goals':22,
        'country':'egypt',
        'club':'liverpool',
        'year':2019,
        },
}
# we make a list of the above dictionary and print for simplicity
player = ['player_2015','player_2016','player_2017','player_2018','player_2019']
print(player)
# we then use the for loop to make some important python work
for player,player_info in players.items():
    print("\nPlayer: " + player)
    full_name = player_info['last'] + ' ' + player_info['first']
    country = player_info['country']
    number_goals = player_info['number_goals']
    club = player_info['club']
    year = player_info['year']
    print('The fullname is: ' + full_name.title())
    # just to know the number of goals scored and what year in one line
    print('Scored ',number_goals,'goals for',club,'in year',year,'to become the highest goal scorer in the EPL that year')

    import matplotlib.pyplot as plt
    plt.scatter(full_name,number_goals,linewidth=5)
    plt.title('Highest goal scorer in the EPL from 2014-2019',fontsize=17)
    plt.xlabel('full_name',fontsize=10)
    plt.ylabel('number_goals',fontsize=10)
    plt.tick_params(axis='both',labelsize=14)
    plt.show()
```
